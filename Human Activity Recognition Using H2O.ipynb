When I try to load train and test datasets, H2O raises H2OResponseError due to duplicate column name. I prevent this by adding a unique number to all colum names.

```python
trainnew = pd.read_csv("train.csv")
testnew = pd.read_csv("test.csv")
```

```python
lst = []
counter = 0
for i in trainnew.columns:
    lst.append(i + str(counter))
    counter += 1
    
trainnew.columns = lst
testnew.columns = lst

trainnew.to_csv("trainnew.csv")
testnew.to_csv("testnew.csv")
```

```python
train = h2o.import_file(path="C://Users//Lenovo//Desktop//Clarusway//DATA SCIENCE//Projects//Human Acitivity Recognition//new.csv")
test = h2o.import_file(path="C://Users//Lenovo//Desktop//Clarusway//DATA SCIENCE//Projects//Human Acitivity Recognition//new2.csv")
```

Indexes are added to dataframes as "C1" column. So that, they will be dropped.

```python
train = train.drop("C1", axis=1)
test = test.drop("C1", axis=1)

print("Shape of Train Dataset:", train.shape)
print("Shape of Test Dataset:", test.shape)
```

I determined features and response columns. Response column is similar target column in sklearn.

```python
features = list(train.columns).remove("Activity562")
response = "Activity562"
```

```python
aml = H2OAutoML(max_models=20, seed=1, nfolds=5, stopping_metric="mean_per_class_error", sort_metric="mean_per_class_error", max_runtime_secs_per_model=300)
aml.train(x=features, y=response, training_frame = train, validation_frame = test)
```

```python
lb = aml.leaderboard
lb.head(rows=lb.nrows)
```

```python
aml.predict(test)
```
